```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import statistics
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
df=pd.read_csv('auto.csv')
df.head()
```

```python
print('Среднее: ', statistics.mean(df['price']))
print('Медиана: ', statistics.median(df['price']))
print('Мода: ', statistics.mode(df['price']))
print('Минимум: ', min(df['price']))
print('Максимум: ', max(df['price']))
print('Среднее отклонение: ', statistics.pstdev(df['price']))
```

```python
df_var = df[['wheel-base', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'city-mpg', 'price']]
df_var.head()
```

```python
#3.МАТРИЦА КОРРЕЛЯЦИЙ
df.corr().style.background_gradient(cmap='coolwarm').set_precision(2)
```

```python
#4.РАСПРЕДЕЛЕНИЕ ПЕРЕМЕННОЙ prise
len_column = len(df['price']) # сколько значений в колонке
# генерируем значения от 0 до len_column 
x = np.linspace(0, len_column, len_column)
# выводим все значения из таблицы
plt.scatter(x, df['price'], marker='.')
```

```python
#5.ВИЗУАЛИЗИРОВАТЬ ВЗАИМОСВЯЗИ
price = df['price']
horsepower = df['horsepower']
# параметр fit_reg - строит линию регрессии для двух выборок
sns.regplot(x= price, y=horsepower, data=df, fit_reg=True)
plt.xlabel('цена')
plt.ylabel('мощность')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
engine_size = df['engine-size']
curb_weight = df['curb-weight']
# параметр fit_reg - строит линию регрессии для двух выборок
sns.regplot(x= engine_size, y=curb_weight, data=df, fit_reg=True)
plt.xlabel('размер двигателя')
plt.ylabel('масса')
plt.title('Корреляционная зависимость двух выборок')
plt.grid(False)
plt.show()
```

```python
bmw_price = df[df['make']=='bmw']['price']
stats.ttest_1samp(bmw_price, df['price'].mean())
```

```python
t-value = 3.909931069594694, мы его сравниваем с табличным значением 2.4469 (в лекционном ноутбуке 5) (alpha = 0.05, N=95). 
Так как значение t-value  превышает 1.45, то мы отвергаем нулевую гипотезу (о том, что средние значения выборок равны). 
Т.е. автомобили БМВ в среднем дороже средней стоимости всех авто.
```

```python
# СРАВНИТЬ цены на audi и volvo
audi = df[df['make']=='audi']['price']
peugot = df[df['make']=='peugot']['price']
print(stats.ttest_ind(audi, peugot))
```
